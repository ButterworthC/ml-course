<h2>Load the data</h2>

```python
pop=pd.read_csv("data/Unit04 Global_Population.csv")
gdp=pd.read_csv("data/Unit04 Global_GDP.csv")

# Define the columns to be discarded
columns_to_discard = [str(year) for year in range(1960, 2001)]

# Drop the columns from 1960 to 2000
pop = pop.drop(columns=columns_to_discard, axis=1)
pop = pop.drop(columns=['2021'])
pop.head()
```

<h2>Get both dataframes to match on Country Code</h2>

```python
# delete all rows in pop where Country Code is NaN

# Delete rows where 'Country Code' is NaN
pop = pop.dropna(subset=['Country Code'])
pop
```

```python
# delete all rows in gdp where Country Code is NaN

gdp = gdp.dropna(subset=['Country Code'])

gdp
```

```python
pop.shape
```

```python
gdp.shape
```

```python
# list and count all the country codes that are in both pop and gdp:

# Perform inner join on 'Country Code' column
merged_df = pd.merge(pop[['Country Code']], gdp[['Country Code']], on='Country Code', how='inner')

# Count and list unique country codes
unique_country_codes = merged_df['Country Code'].unique()
num_country_codes = len(unique_country_codes)

# Print unique country codes and count
print("Unique Country Codes:", unique_country_codes)
print("Number of Country Codes:", num_country_codes)
```

```python
# list and count all the country codes that are in pop (267 rows) but not in gdp (266 rows):

# Perform left join between pop and gdp on 'Country Code'
merged_df = pd.merge(pop[['Country Code']], gdp[['Country Code']], on='Country Code', how='left', indicator=True)

# Filter out rows where 'Country Code' from gdp is null (indicating no match)
rows_not_in_gdp = merged_df[merged_df['_merge'] == 'left_only']

# List and count unique country codes
unique_country_codes = rows_not_in_gdp['Country Code'].unique()
num_country_codes = len(unique_country_codes)

# Print unique country codes and count
print("Unique Country Codes:", unique_country_codes)
print("Number of Country Codes:", num_country_codes)
```

```python
# perhaps the extra one in pop is a duplicate
# List all the duplicate Country Codes in pop

# Filter out rows with duplicate Country Codes
duplicate_country_codes = pop[pop.duplicated(subset='Country Code', keep=False)]

# List unique duplicate Country Codes
unique_duplicate_country_codes = duplicate_country_codes['Country Code'].unique()

# Print the list of duplicate Country Codes
print("Duplicate Country Codes in pop:", unique_duplicate_country_codes)
```

```python
# show every row in pop where Country Code = 'ABW'

# Filter rows where 'Country Code' is 'ABW'
rows_abw = pop[pop['Country Code'] == 'ABW']

# Display the rows
print(rows_abw)
```

```python
# Display Country Name and Country Code columns from index 0 to 13 inclusive
subset_0_to_13 = pop.loc[0:13, ['Country Name', 'Country Code']]

# Display the subset
print(subset_0_to_13)
```

```python
# Delete the zeroth row in pop
pop = pop.drop(0)

# Display Country Name and Country Code columns from index 0 to 13 inclusive
subset_0_to_13 = pop.loc[0:13, ['Country Name', 'Country Code']]

# Display the subset
print(subset_0_to_13)
```

<h2>Investigate any correlation between the mean population of each country and the mean per capita GDP (from 2001 to 2021). Very briefly, interpret the generated plot. Evaluate the Pearson Correlation Coefficient.</h2>

```python
# add mean columns to pop and gdp
# Select columns from '2001' to '2020'

selected_columns = pop.loc[:, '2001':'2020']

# Convert non-numeric values to NaN
selected_columns = selected_columns.apply(pd.to_numeric, errors='coerce')

# Calculate the sum of values across selected columns
sum_values = selected_columns.sum(axis=1)

# Count the number of non-NaN values in each row
count_non_nan = selected_columns.notnull().sum(axis=1)

# Calculate the mean_pop column
pop['mean_pop'] = sum_values / count_non_nan

# Display the DataFrame with the new column
print(pop)

```

```python
# same for gdp:

# Select columns from '2001' to '2020'

selected_columns = gdp.loc[:, '2001':'2020']

# Convert non-numeric values to NaN
selected_columns = selected_columns.apply(pd.to_numeric, errors='coerce')

# Calculate the sum of values across selected columns
sum_values = selected_columns.sum(axis=1)

# Count the number of non-NaN values in each row
count_non_nan = selected_columns.notnull().sum(axis=1)

# Calculate the mean_pop column
gdp['mean_gdp'] = sum_values / count_non_nan

# Display the DataFrame with the new column
print(gdp)
```

```python
# scatter plot of pop and gdp...
# Calculate the mean GDP for each row
# gdp['mean_gdp'] = gdp.loc[:, '2001':'2020'].mean(axis=1)

# Create scatter plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(pop['mean_pop'], gdp['mean_gdp'], alpha=0.5)
plt.title('Scatter Plot of Mean Population vs Mean GDP')
plt.xlabel('Mean Population')
plt.ylabel('Mean GDP')
plt.grid(True)
plt.show()
```

```python
# let's pretend it's OK and work out the Pearson Correlation Coefficient

# Calculate the Pearson correlation coefficient
pearson_corr = pop['mean_pop'].corr(gdp['mean_gdp'])

# Display the result
print("Pearson Correlation Coefficient between mean_pop and mean_gdp:", pearson_corr)
```
